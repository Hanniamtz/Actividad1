```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel("Datos de Facturación.xlsx")
df
```

CVE_CLPV (valores: 1000 a 2000)

```python
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')
flt1 = df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
flt1
```

```python
flt1.to_csv("flt1.csv")
from google.colab import files 
files.download("flt1.csv")
```
