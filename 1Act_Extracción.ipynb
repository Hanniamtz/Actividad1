#Filtro por objeto

#Filtro por comparacion


```python
#Muestra las variables en donde el costo unitario es menor que cien
filtro6=df[df["COSTO_UNITARIO"] < 100]
filtro6
```

```python
filtro6.to_csv("Filtro6.csv")
from google.colab import files 
files.download("Filtro6.csv")
```

#Filtro y cumplir dos condiciones 

```python
#Se muestra las filas en donde el precio unitario es mayor a cien y fueron realizadas en la siguiente fecha 2022-01-05
filtro7=df[(df["PRECIO_UNITARIO"] > 100) & (df["FECHA_DOC"] == "2022-01-05")]
filtro7
```

```python
filtro7.to_csv("Filtro7.csv")
from google.colab import files 
files.download("Filtro7.csv")
```

#Filtro o con que se cumpla una de las dos condiciones 

```python
#En este se muestra filtrada las opciones con que se cumpla una de las dos condiciones  ya sea que el MARGEN_UNITARIO_CALCULADO sea menor que 1000 
#o que la fecha del doc sea en 2022-01-05 
filtro8=df[(df["MARGEN_UNITARIO_CALCULADO"] < 1000) | (df["FECHA_DOC"] == "2022-01-05")]
filtro8
```

```python
filtro8.to_csv("Filtro8.csv")
from google.colab import files 
files.download("Filtro8.csv")
```

#Filtro not

```python
#excluye los siguientes datos en donde el MARGEN_UNITARIO_CALCULADO es mejor o igual a mil y que el nombre del cliente no sea JUAN ANTONIO RESENDIZ UGALDE
filtro9=df[~(df["MARGEN_UNITARIO_CALCULADO"] <= 1000) & ~(df["NOMBRE_CLIENTE"] == "JUAN ANTONIO RESENDIZ UGALDE")]
filtro9
```

```python
filtro9.to_csv("Filtro9.csv")
from google.colab import files 
files.download("Filtro9.csv")
```
